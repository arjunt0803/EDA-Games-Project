```python
import pandas as pd                  
import matplotlib.pyplot as plt     
import seaborn as sns               
```

```python
df = pd.read_csv('clean_dates.csv')      
 
print("\n✅ Dataset loaded!")
print("Total rows and columns:", df.shape)  
 
```

```python
print("\n--- First 5 rows ---")
print(df.head())                    
 
print("\n--- Column names ---")
print(df.columns.tolist())           
 
print("\n--- Data types ---")
print(df.dtypes)                    
 
print("\n--- Missing values ---")
print(df.isnull().sum())            
 
```

```python
df = df.dropna(subset=['Year', 'Publisher'])
 
df['Year'] = df['Year'].astype(int)
 
df = df[df['Year'] <= 2020]
 
print("\n✅ Data cleaned!")
print("Rows remaining:", len(df))
```

```python
top10 = df.nlargest(10, 'Global_Sales')[['Name', 'Global_Sales']]
 
print("\n--- Top 10 Best Selling Games ---")
print(top10.to_string(index=False))
 
plt.figure(figsize=(10, 6))
plt.barh(top10['Name'], top10['Global_Sales'], color='steelblue')
plt.xlabel('Global Sales (Millions)')
plt.title('Top 10 Best Selling Games of All Time')
plt.gca().invert_yaxis()             
plt.tight_layout()
plt.savefig('chart1_top10_games.png')
plt.show()
```

```python
top10 = df.nlargest(10, 'Global_Sales')[['Name', 'Global_Sales']]
 
print("\n--- Top 10 Best Selling Games ---")
print(top10.to_string(index=False))
 
plt.figure(figsize=(10, 6))
plt.barh(top10['Name'], top10['Global_Sales'], color='steelblue')
plt.xlabel('Global Sales (Millions)')
plt.title('Top 10 Best Selling Games of All Time')
plt.gca().invert_yaxis()             
plt.tight_layout()
plt.savefig('chart1_top10_games.png')
plt.show()
```

```python
genre_sales = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
 
print("\n--- Sales by Genre ---")
print(genre_sales)
 
plt.figure(figsize=(10, 6))
genre_sales.plot(kind='bar', color='coral')
plt.xlabel('Genre')
plt.ylabel('Total Sales (Millions)')
plt.title('Which Game Genre Sells the Most?')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart2_genre_sales.png')
plt.show()
```

```python
platform_sales = df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).head(10)
 
print("\n--- Top 10 Platforms ---")
print(platform_sales)
 
plt.figure(figsize=(10, 6))
platform_sales.plot(kind='bar', color='mediumseagreen')
plt.xlabel('Platform')
plt.ylabel('Total Sales (Millions)')
plt.title('Top 10 Gaming Platforms by Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart3_platform_sales.png')
plt.show()
```

```python
yearly = df.groupby('Year')['Global_Sales'].sum()
 
print("\n--- Yearly Sales ---")
print(yearly)
 

plt.figure(figsize=(12, 5))
plt.plot(yearly.index, yearly.values, color='purple', marker='o')
plt.xlabel('Year')
plt.ylabel('Total Sales (Millions)')
plt.title('Video Game Sales Over the Years')
plt.tight_layout()
plt.savefig('chart4_yearly_sales.png')
plt.show()
```

```python
na    = df['NA_Sales'].sum()
eu    = df['EU_Sales'].sum()
jp    = df['JP_Sales'].sum()
other = df['Other_Sales'].sum()
 
regions = [na, eu, jp, other]
labels  = ['North America', 'Europe', 'Japan', 'Others']
colors  = ['#3498db', '#2ecc71', '#e74c3c', '#f39c12']
 
plt.figure(figsize=(7, 7))
plt.pie(regions, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Which Region Buys the Most Games?')
plt.tight_layout()
plt.savefig('chart5_regional_pie.png')
plt.show()
```

```python
top_publishers = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
 
print("\n--- Top 10 Publishers ---")
print(top_publishers)
 
plt.figure(figsize=(10, 6))
top_publishers.plot(kind='bar', color='goldenrod')
plt.xlabel('Publisher')
plt.ylabel('Total Sales (Millions)')
plt.title('Top 10 Game Publishers by Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('chart6_publishers.png')
plt.show()
```

```python
print("\n" + "="*45)
print("        FINAL SUMMARY")
print("="*45)
print(f"Total Games          : {len(df):,}")
print(f"Total Global Sales   : {df['Global_Sales'].sum():.0f} Million")
print(f"Best Selling Game    : {df.loc[df['Global_Sales'].idxmax(), 'Name']}")
print(f"Most Popular Genre   : {genre_sales.idxmax()}")
print(f"Top Platform         : {platform_sales.idxmax()}")
print(f"Top Publisher        : {top_publishers.idxmax()}")
print(f"Peak Sales Year      : {yearly.idxmax()}")
print("="*45)
```
